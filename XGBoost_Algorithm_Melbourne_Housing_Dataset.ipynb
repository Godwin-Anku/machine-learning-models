```python
import numpy as np
import pandas as pd
```

```python
X_train_full_data = pd.read_csv('/content/train (1).csv')
X_test_full_data = pd.read_csv('/content/train (1).csv')
```

```python
X_train_full_data
```

```python
X_test_full_data
```

**Removing missing rows with reference to target values**

```python
def row_dropping(data):
  data.dropna(axis = 0, subset = ['SalePrice'], inplace = True)
  return data
```

```python
X_train_full_data = row_dropping(X_train_full_data)
X_train_full_data
```

```python
X_test_full_data = row_dropping(X_test_full_data)
X_test_full_data
```

```python
y_train_full_data = X_train_full_data.SalePrice
y_test_full_data = X_test_full_data.SalePrice
```

```python
y_train_full_data
```

```python
y_test_full_data
```

**Numerical columns**

```python
numerical_cols = [cname for cname in X_train_full_data.columns if X_train_full_data[cname].dtype in ['int64', 'float64']]
print(len(numerical_cols))
numerical_cols
```

**Categorical columns**

```python
low_cardinality_cols = [cols for cols in X_train_full_data.columns if X_train_full_data[cols].dtype == 'object' and X_train_full_data[cols].nunique() < 10]
print(len(low_cardinality_cols))
low_cardinality_cols
```

```python
#Selected columns

selected_cols = numerical_cols + low_cardinality_cols
X_train_full_data = X_train_full_data[selected_cols]
X_test_full_data = X_test_full_data[selected_cols]
```

```python
X_train_full_data
```

```python
X_test_full_data
```

**Numerical Imputation**

```python
from sklearn.impute import SimpleImputer


def numerical_imputation(num_strategy, X_full_data, selected_data):
  num_imputer = SimpleImputer(strategy = num_strategy)
  X_data_im = pd.DataFrame(num_imputer.fit_transform(X_full_data[selected_data]))
  X_data_im.columns = selected_data
  return X_data_im
```

```python
X_train_numerical_data = numerical_imputation('most_frequent', X_train_full_data, numerical_cols)
X_train_numerical_data
```

```python
X_test_numerical_data = numerical_imputation('most_frequent', X_test_full_data, numerical_cols)
X_test_numerical_data
```

**Categorical Imputation and One Hot Encoding**

```python
from sklearn.preprocessing import OneHotEncoder

def categorical_trans(cat_strategy, X_data, selected_data):
  cat_imputer = SimpleImputer(strategy = cat_strategy)
  X_cat_data_im = pd.DataFrame(cat_imputer.fit_transform(X_data[selected_data]))
  X_cat_data_im.columns = selected_data
  OH_encoder = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
  X_cat_OH = pd.DataFrame(OH_encoder.fit_transform(X_cat_data_im))
  X_cat_OH.index = X_cat_data_im.index
  return X_cat_OH
```

```python
X_fulltrain_categorical_data = categorical_trans('most_frequent', X_train_full_data, low_cardinality_cols)
X_fulltrain_categorical_data
```

```python
X_fulltest_categorical_data = categorical_trans('most_frequent', X_test_full_data, low_cardinality_cols)
X_fulltest_categorical_data
```

**Concatenate all transformed columns**

```python
X_final_train = pd.concat([X_train_numerical_data, X_fulltrain_categorical_data], axis = 1)
X_final_test = pd.concat([X_test_numerical_data, X_fulltest_categorical_data], axis = 1)
```

```python
X_final_train
```

```python
X_final_test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X_final_train, y_train_full_data, train_size = 0.7, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
X_valid
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators = 1000, learning_rate = 0.05, n_jobs = -1)
model.fit(X_train, y_train, early_stopping_rounds = 5, eval_set = [(X_valid, y_valid)], verbose = False)
```

```python
y_predict = model.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_valid, y_predict)
print(f'MAE: {mae}')
```

```python
y_test_predict = pd.DataFrame(model.predict(X_final_test), columns = ['Predicted SalePrice'])
y_test_predict
```
