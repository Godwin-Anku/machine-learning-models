**Columns with missing values**

**Numeric columns with missing values**

```python
numeric_missing_columns = [col_name for col_name in missing_values if X_train[col_name].dtype in ['int64', 'float64']]
numeric_missing_columns
```

```python
categorical_missing_columns = set(missing_values) - set(numeric_missing_columns)
categorical_missing_columns
```

**Imputation for numeric data and One-Hot Encoding of Categorical Data**

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

def apply_imputer(data, imputation_method, X_train):
  imputer = SimpleImputer(strategy = imputation_method)
  imputed_data = pd.DataFrame(imputer.fit_transform(X_train[data]))
  imputed_data.columns = data
  return imputed_data
```

```python
new_numeric_data = apply_imputer(numeric_missing_columns, 'mean', X_train)
```

```python
new_numeric_data.head(50)
```

**Imputation of Categorical Dataset**

```python
new_categorical_data = apply_imputer(categorical_missing_columns, 'most_frequent', X_train)
new_categorical_data.head(50)
```

**Applying One-Hot Encoding to Categorical Dataset**

```python
def OH_encoding(data):
  object_columns = data.select_dtypes(include = 'object').columns
  if not object_columns.empty:
    OH_encoder = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
    OH_cols_data = pd.DataFrame(OH_encoder.fit_transform(data))
    OH_cols_data.index = data.index
    OH_cols_data.columns = OH_cols_data.columns.astype(str)
    return OH_cols_data
  else:
    return object_columns
```

```python
final_categorical_data = OH_encoding(new_categorical_data)
final_categorical_data
```

**Concatenating numerical data and categorical data**

```python
X_train_full = pd.concat([final_categorical_data, new_numeric_data], axis = 1)
X_train_full
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train_full, y_train_full, train_size = 0.7, test_size = 0.3, random_state = 0)
print(f'X train data : \n{X_train}')
print(f'X test data : \n{X_test}')
print(f'Y train data : \n{y_train}')
print(f'Y test data : \n{y_test}')
```

```python
model = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0)
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = pd.DataFrame(model.predict(X_test), columns = ['Predicted SalePrice'])
```
